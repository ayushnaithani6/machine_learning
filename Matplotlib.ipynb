# matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setop plot
fig, ax = plt.subplots(figsize=(10, 5)) # width and height

# 3. Plot the data 
ax.plot(x, y)

# 4. customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel = "y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig('./images/sample-plot.png')
```

```python
# Histogram

# Make some data
x = np.random.randn(1000)

# plot the data
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(x);
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5));

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(x);
```

# Plotting from pandas dataframe 
object oriented method

### Which one should you use? (pyplot vs matplotlib OO method)

* when plotting something quickly its okay to use pyplot method
* when plotting something advance use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", 
             x = "age",
             y = "chol",
             c = "target");
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);

# limitations of x and y axis 
ax.set_xlim([45, 100]);   # set x axis from 45 to 100
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize the plot
ax.set(title="Heart disease and Cholestrol levels",
       xlabel="Age",
       ylabel="Cholestrol");

# Add a legend
# * means unpack all the values
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), 
           linestyle="--");

ax.axvline(over_50["age"].mean(),
           linestyle="--");
```

```python
over_50.head()
```

```python
# Working with subplots

# Create figures and axes
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)

#### plot the data on subplot1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"]);

# customize the subplot1
ax1.set(title="heart disease and cholestrol levels",
        ylabel="cholestrol")

# add legend to subplot2
ax1.legend(*scatter1.legend_elements(), title="Target")

# add horizontal line passing through mean of cholestrol y-axis
ax1.axhline(y=over_50["chol"].mean(),
            linestyle="--")



#### plot the data on subplot2
scatter2 = ax2.scatter(x=over_50["age"],
                       y=over_50["thalach"], 
                       c=over_50["target"])

# customize subplot2
ax2.set(title="heart disease and Max Heart rate",
        xlabel="age",
        ylabel="max heart rate");

# add legend to subplot2
ax2.legend(*scatter2.legend_elements(), title="Target");

# add horizontal line passing through mean of thalach
ax2.axhline(y=over_50["thalach"].mean(),
            linestyle="--")


# add a title to entire figure
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");
```

```python
# customizing x and y axis limitations

# Create figures and axes
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)

#### plot the data on subplot1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"],
                       cmap="winter");

# customize the subplot1
ax1.set(title="heart disease and cholestrol levels",
        ylabel="cholestrol")

# set xlim and ylim
ax1.set_xlim([50, 80])

# add legend to subplot2
ax1.legend(*scatter1.legend_elements(), title="Target")

# add horizontal line passing through mean of cholestrol y-axis
ax1.axhline(y=over_50["chol"].mean(),
            linestyle="--")



#### plot the data on subplot2
scatter2 = ax2.scatter(x=over_50["age"],
                       y=over_50["thalach"], 
                       c=over_50["target"],
                       cmap="winter")

# customize subplot2
ax2.set(title="heart disease and Max Heart rate",
        xlabel="age",
        ylabel="max heart rate");

# change xlim and y lim
ax2.set_xlim([50, 80])
ax2.set_ylim([60, 200])

# add legend to subplot2
ax2.legend(*scatter2.legend_elements(), title="Target");

# add horizontal line passing through mean of thalach
ax2.axhline(y=over_50["thalach"].mean(),
            linestyle="--")


# add a title to entire figure
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold");
```
